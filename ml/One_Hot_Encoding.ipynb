```python
import pandas as pd

# Create a sample DataFrame
data = {'color': ['red', 'green', 'blue', 'red']}
df = pd.DataFrame(data)

# Perform one-hot encoding using get_dummies
encoded_df = pd.get_dummies(df, columns=['color'], prefix=['color'])

# Display the encoded DataFrame
print(encoded_df) # To see the output, run the code.
```

# One Hot Encoder Using Sklearn

```python
data = d = {'sales': [100000,222000,1000000,522000,111111,222222,1111111,20000,75000,90000,1000000,10000],
            'city': ['Tampa','Tampa','Orlando','Jacksonville','Miami','Jacksonville','Miami','Miami','Orlando','Orlando','Orlando','Orlando'],
            'size': ['Small', 'Medium','Large','Large','Small','Medium','Large','Small','Medium','Medium','Medium','Small',]
            }

df = pd.DataFrame(data)

df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(handle_unknown='ignore')

ohe_df = pd.DataFrame(ohe.fit_transform(df[['city']]).toarray())
ohe_df
```
